```python
import matplotlib.pyplot as plt
```

```python
class1_x_value = [0,1,2,3,4]
class1_y_value = [1,3,5,7,9]
class1_x_value, len(class1_x_value), class1_y_value, len(class1_y_value)
```

```python
plt.scatter(x=class1_x_value, y=class1_y_value)
```

```python
class2_x_value = [1.5,2.5,3.5,4.5,5.5]
class2_y_value = [1.5,3.5,6.5,7.5,8.5]
class2_x_value, len(class2_x_value), class2_y_value, len(class2_y_value)
```

## 하나 plot 표시

```python
# 두 클래스 변수를 한 plot에 표현
plt.scatter(x=class1_x_value, y=class1_y_value)
plt.scatter(x=class2_x_value, y=class2_y_value)
plt.show()
```

### title, label, legend(제목, 레이블, 범례)

```python
plt.title('Classifications Plot')
plt.scatter(x=class1_x_value, y=class1_y_value, label='Class One')
plt.scatter(x=class2_x_value, y=class2_y_value, label='Class Two')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.legend()
plt.show()
```
