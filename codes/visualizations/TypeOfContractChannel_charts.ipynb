#### 단일변수 범주형 : 주로 문자, 항목

```python
df_TOCC['type_of_contract'].value_counts()
```

```python
sns.countplot(data=df_TOCC, x='type_of_contract') # 단일변수의 빈도수를 보고 싶을 때
plt.title("계약 종류")
plt.show()
```

```python
df_TOCC['type_of_contract2'].value_counts()
```

```python
sns.countplot(data=df_TOCC, x='type_of_contract2', hue='product') #다변수 일 때
plt.title("계약 종류 2") 
plt.show()
```

```python
sns.countplot(data=df_TOCC, x='type_of_contract2')
```

```python
sns.countplot(data=df_TOCC, y='type_of_contract2')
```

```python
df_TOCC['bank'].value_counts()
```

```python
sns.countplot(data=df_TOCC, y='bank')
plt.show()
```

```python
# 거래 은행 중 상위 7개 확인 
# df_TOCC.value_counts('bank')
li_bank_top7=df_TOCC['bank'].value_counts().index.to_list()[:7]
print(li_bank_top7)
```

```python
# sns.countplot(data=df_TOCC, y='bank', order=['국민은행', '롯데카드', '농협은행'])
sns.countplot(data=df_TOCC, y='bank', order=li_bank_top7)
plt.show()
```
